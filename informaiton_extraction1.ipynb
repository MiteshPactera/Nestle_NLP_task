```python
import spacy
import pandas as pd
pd.set_option('display.max_colwidth',None)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_excel('test_sheet2.xlsx')
```

```python
df.head()
```

```python
for i in range(len(df)):
    print(df['Sample General English Query:'][i])
```

```python
nlp_text_df = {}
for i in range(len(df)):
    nlp_text_df[i] = nlp(df['Sample General English Query:'][i])
```

```python
# Let's exract information from the above sentence
numbers_dict = {}
feats_and_time_dict = {}
intent_dict = {}

for i in range(len(nlp_text_df)):
    num = []
    f_t = []
    vrb = []
    for token in nlp_text_df[i]:
        if token.pos_ == 'NUM':
            num.append(token)
      
        if token.pos_ =='PROPN' or token.pos_ == 'NOUN':
            f_t.append(token)
        
        if token.pos_ == 'VERB':
            vrb.append(token)
    
    numbers_dict[i] = num
    feats_and_time_dict[i] = f_t
    intent_dict[i] = vrb
```

```python
#print(numbers_dict,'\n\n', feats_and_time_dict, '\n\n', intent_dict)
```

```python
df['Features_info'] = feats_and_time_dict.values()
df['Numeric_info'] = numbers_dict.values()
df['Intent'] = intent_dict.values()
```
